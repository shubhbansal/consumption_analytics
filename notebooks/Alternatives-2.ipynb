## Finding alternatives to packaged nuts 

- Find the closest matching product names - this allows cashews to be suggested for cashew and peanuts for peanuts
- Find the price difference (per gram) basis and suggest the ones with the smallest difference

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
import seaborn as sns
import difflib
%matplotlib inline 
```

```python
engine = create_engine('mysql://root:root@localhost:3306/ptdc_divish')
```

```python
dfp = pd.read_sql('select * from product_summary', engine, index_col='index')
del dfp['Unnamed: 0']
```

# Nuts

```python
# dfb = dfp[(dfp['meta_category_description'] =='Biscuits & Snacks') & (dfp['product_description'].str.contains('BISC', case=False))]

dfn = dfp[(dfp['meta_category_description'] =='Nuts')]
```

```python
non_nuts = ['butter', 'oil', 'discon', 'milk', 'toffee', 'jar', 'discon', 'syrup']
"|".join(non_nuts)
```

```python
dfn.head()
```

```python
dfn_loose = dfn[dfn['measurement_unit']<>'Pk'][['product_description', 'category_description',
                                                'avg_price']].drop_duplicates(['product_description', 'category_description'])

dfn_loose = dfn_loose[~dfn_loose['product_description'].str.contains("|".join(non_nuts), case=False)]
dfn_loose['weight'] = dfn_loose['product_description'].str.extract('(\d+)')
dfn_loose.loc[dfn_loose['product_description'].str.contains('kg', case=False), 'weight'] = 1000
```

```python
dfn_loose.head()
```

** Filling weight fo cashew, per 100gm price is around Rs 100 **

```python
cond = (dfn_loose['product_description'].str.contains('cashew', case=False)) & (dfn_loose['weight'].isnull())
dfn_loose.loc[cond, 'weight'] = 100
```

```python
dfn_loose.tail()
```

**Filling in NAN values for dehydrated fruits. It should probably be per kg price **

```python
cond = (dfn_loose['product_description'].str.contains('dehydrated|dried', case=False)) & (dfn_loose['weight'].isnull())
dfn_loose.loc[cond, 'weight'] = 1000
```

```python
dfn_loose.head()
```

```python
dfn_loose['weight'] = dfn_loose['weight'].astype(np.float)
```

```python
dfn_loose['price/weight'] = dfn_loose['avg_price']/dfn_loose['weight']
```

### Packed Goods 

```python
dfn_pack = dfn[dfn['measurement_unit']=='Pk'][['product_description', 'category_description',
                                                'avg_price']].drop_duplicates(['product_description', 'category_description'])

dfn_pack = dfn_pack[~dfn_pack['product_description'].str.contains("|".join(non_nuts), case=False)]

# Extracting weights
dfn_pack['weight'] = dfn_pack['product_description'].str.extract('(\d+)')
dfn_pack.loc[dfn_pack['product_description'].str.contains('kg', case=False), 'weight'] = 1000
```

```python
dfn_pack.head()
```

** Filling in missing weight values for cashew **

```python
cond = (dfn_pack['product_description'].str.contains('cashew', case=False)) & (dfn_pack['weight'].isnull())
dfn_pack.loc[cond,'weight'] = 100
```

```python
dfn_pack['weight'] = dfn_pack['weight'].astype(np.float)
dfn_pack['price/weight'] = dfn_pack['avg_price']/dfn_pack['weight']
```

```python
dfn_pack.head()
```

```python
recommend_df = pd.DataFrame(columns={'recommendation', 'price_difference'})

for row in dfn_pack.iterrows():
  
    temp = dfn_loose.copy() 
    alt = difflib.get_close_matches(row[1]['product_description'].strip(),temp['product_description'].str.strip().tolist(),
                                             cutoff=0.4)
    recommend_df.loc[row[1]['product_description'],'recommendation'] = alt
    
    diff = []
    for a in alt:
        price1 = dfn_pack.loc[dfn_pack['product_description']==row[1]['product_description'], 'price/weight'].values[0]
        price2 = dfn_loose.loc[dfn_loose['product_description']==a, 'price/weight'].values[0]
        
        diff.append(price1-price2)
    
    recommend_df.loc[row[1]['product_description'],'price_difference'] = diff
                         
    
        
```

```python
recommend_df.head()
```

```python
best_recommend = recommend_df['recommendation'].apply(pd.Series)
```
